## <font color='Green'>1.1. Karar Ağacı Terminolojisi

### Öncelikle terminolojiyi biraz ele alalım.




| Kök düğüm (root node) :|Tüm örneklemi içeren ilk düğümdür.   |
|------|------|
|  İç Düğüm/Karar Düğümü (decision node) :| Altında daha başka düğümler içeren düğümdür.|
|------|------|
| Yaprak Düğüm / Son Nokta (leaf node/terminal node) :|Altında düğüm olmayan düğümlerdir. |
|------|------|
| Bölme (splitting) :|  Bir düğümü alt dallara ayırma işlemidir.|
|------|------|
|Budama (pruning) : |Bir düğümün altındaki düğümlerin ağaçtan çıkarılmasıdır. |
|------|------|
| Dal/Alt Ağaç (branch) : |Bir düğüm ve onun altındaki tüm kısımlardır.  |
|------|------|
| Üst Düğüm (parent node) : | Bir düğüm hiyerarşik olarak bir üstünde bulunan düğümdür. |
|------|------|
|Alt/Çocuk Düğüm (child node) :|  Bir düğümün kendisine bağlı olan düğümlerdir.|
|------|------|
| Düğüm (node):| Yukarıdeki örnekte her bir soru bir düğüm (node)'dür.Düğümleri kök (temel) düğüm, iç düğüm (takip eden düğümler) ve yaprak düğüm (son nokta) olarak ayırabiliriz.|



Terim akışına koyduğunuzda ilk düğümden başlarsınız, bir yaprak düğümüne varana kadar iç düğümlere doğru dalları takip edersiniz. Bütün bu olay bir ağaca benzetilebilir, bir ağacı ters çevirirseniz böylece kökleri aşağıda dallar ve yapraklar yukarı kısımda olacaktır.

Her kural veriyi belirli alt gruplara böler, tipik olarak özellikle "evet veya hayır" soruları yaygındır. Ardı ardına sorularla sonuca doğru ulaşmaya çalışırız. Burada en önemli husulardan biri ağacı en iyi şekilde ayıracak soruyu sorabilmekte yatmaktadır.

## <font color='green'> 1.2. Karar Ağaçlarının kullanım alanları, avantaj ve dezavantajları

#### Kullanım Alanları

* **Karar ağaçları** görsel bir sunum sağladığından bir veri üzerinden karar vermeye de katkı sağlamaktadır. Hatta yapılacak küçük değişikliğin bile kararı nasıl etkilediğini gösterebilmektedir.

* Veri kümesinde **çok fazla hatalı veri ya da eksik veri varsa** karar ağaçları tercih edilebilir.

* **Veri kümesindeki hedef değişken** discrete değişken ise yanı sayılabilir bir veri ise karar ağaçları için daha uygundur.

* Karar ağaçları **finans alanında** yaygın olarak kullanılmaktadır.

#### Avantajları

* **Anlaşılması kolaydır**. Ağaç şeklinde olan bir gösterim ile, istatistik bilgisi olmayan kişiler de veriyi kolaylıkla yotumlayabilir.

* Veri keşfinde çok kullanışlıdır. Yüzlerce değişkeni olan bir veri kümesinde en önemli değişkenleri ortaya çıkarabilir. **Hedef değişkeni etkileyen yeni değişkenler üretmekte de kolaylık sağlar**.

* Veri temizlemesine diğer algoritmalara nazaran daha az ihtiyaç duyulur. **Eksik değerler ya da aykırı değerler modeli pek etkilemez**.

* Veri tipinin bir önemi yoktur. **Numerik ya da kategorik olabilir**.

* **Ayarlanacak parametreler çok azdır**.

#### Dezavantajları

* **Overfitting'e açıktır**. Dalların budanması ile çözülebilir.

* Sürekli numerik değişkenler ile çalışırken, **verileri kategorikleştirdiği için veri kaybı yaşanabilmektedir**.

* Veri kümesindeki yer alan küçük farklılıklar tamamen farklı bir ağaç üretilmesine neden olabilir.

* **Veri kümesi dengesiz ise bir tarafa doğru yanlı bir ağaç** ortaya çıkabilir.

* **Modelin eğitimi uzun sürebilir**.

## <font color='green'> 1.3. Karar Ağacı Algoritması Nasıl Çalışır?

Karar ağacı algoritması problemi verileri ağaç şeklinde temsil ederek çözmeye çalışır. Her karar düğümü bir değişkene ve her yaprak düğümü de hedef etiketine karşılık gelir. Ağaç oluşturulurken şu sıra takip edilir.

* Ağacın köküne en uygun değişken koyulur. Mümkün olduğunca basit olmalıdır.

* Veri kümesi alt kümelere bölünür.

* Yaprak düğümlere ulaşıncaya yukarıdaki işlemlere devam edilir.

## <font color= 'green'> 2. Karar Ağaçları ile Sınıflandırma ve Regresyon

Karar ağaçlarını, regresyon ağaçları ve sınıflandırma ağaçları olarak ikiye ayırabiliriz. Genel olarak baktığımızda her iki ağaç yöntemi de benzer şekilde çalışmaktadır.

* Regresyon ağaçlarında son düğüm tarafından elde edilen değer, o bölgeye denk gelen verilerin ortalama değeridir. Dolayısıyla bilinmeyen bir veri bu bölgeye düşerse tahmin bu ortalama değere göre yapılmaktadır.

* **Sınıflandırma ağaçlarında son düğüm tarafından elde edilen değer, o bölgeye denk gelen sınıfların mod değeridir.** Dolayısıyla bilinmeyen bir veri bu bölgeye düşerse tahmin bu mod değere göre yapılmaktadır.

* Hem sınıflandırma hemde regresyon ağaçlarında bölgeler birbirinden ayrıdır ve **çakışma olmaz.**

* Ağaç dallara ayrılırken o düğüm için en başarılı ayrım seçilir. İleride ki dallanmalar dikkate alınmaz.

* **Ağaçların dallara göre bölünmesi, belirleyeceğimiz dal sayısına kadar devam eder**. Bu değer iyi belirlenmezse *overfitting* olabilir.

## <font color='green'>2.1. Ağacın nasıl bölüneceği nasıl belirlenecek?
    
    
Ağacın nasıl bölüneceği karar ağaçlarının en önemli noktasıdır ve ağacın doğruluğunu çok fazla etkilemektedir. Bölerken dikkat edilen en önemli kriter alt dalda en fazla homojenliği sağlayabilmek ve dalın hedef değişkenler açısından saf olmasını sağlamaktır.

Bir karar ağacı tasarlarken mümkün olduğunca en verimli olanı isteriz. Bütün soruların aynı miktarda bilgiye sahip olmadığını beklemek mantıklıdır. Yukarıdaki örneğe bakacak olursak, Onlara Norveç’te olup olmadığını sorarsak, epeyce kullanışsız bir soru olacaktır. Onların cevaplarından hiçbir bilgi edinemeyeceğiz.

Soruların seçiciliğini sınıflandırma ağaçlarında değerlendirirken iki yöntem kullanılmaktadir.

* **Bilgi kazanımı (information gain) ve Düzensizlik (entropy).**
* **Gini Index**

Regresyon ağaçlarında ise Ortalama Kök Hata (MSE) ağacın dallara ayrılmasında kriter olarak kullanılır.

### <font color='green'> 2.1.1. Düzensizlik (Entropy)
    
*Entropy (düzensizlik)* diğer bilim disiplinlerinden tanışık olabileceğiniz bir terimdir, düzensizlik, karışıklık, bozukluk anlamı vardır. Karar ağaçlarında kullanımına baktığımızda ağacın nasıl dallanacığını dolayısıyla karar ağacının sınırlarını belirler. Düzensizliği bilgi kazanımını ölçmek için kullanabiliriz, başlangıç durumu ile takip eden durumların ağırlıklı olası çıktılarının değişimi düzensizlik olarak tanımlanır.

Bu denklem olası bütün çıktıların toplamlarının ağırlıklı ortalamasından gelir. Buradaki önemli husus, bunun çıktıların içindeki belirsizliğinin ölçümü olmasıdır. Olası çıktı sayısını azalttığımızda ve çıktılara daha fazla güvendiğimizde düzensizlik azalacaktır. Ağacın bir bölgesinde olası bir çıktı/dal varsa düzensizlik sıfır olacaktır. Çünkü belirsizlik olmayacaktır.

Düzensizlik değerini bilgi kazanımını hesaplamak için kullanacağız.

Etkili bir karar ağacı modeli dizayn etmenin metodu, mümkün olduğunca hızlı şekilde en çok bilgiyi toplayabilmektir. Ne kadar bilgi toplabildiğimiz ise **Bilgi Kazanımı (IG)** ile ölçülmektedir. Karar ağacı algoritması her zaman bilgi kazanımını artırmak için çabalar ve en yüksek bilgi kazanımına göre ağacı dallara ayırır.

### <font color='green'>2.1.2. Gini Index
Ağacın dallara ayrılmasında mümkün olduğunca saf düğümler oluşturulmasının en önemli kriterler olduğunu belirtmiştik. Bu saflığı ölçmenin diğer yöntemi de Gini Index'i kullanmaktır.

```python
X = df2.drop('Life_Expectancy', axis=1)
y = df2.Life_Expectancy
```

```python
X_eğitim, X_test, y_eğitim, y_test =  train_test_split(X, y, test_size=0.20, random_state=40)
```

```python
from sklearn.metrics import mean_squared_error as MSE
```

```python
y_tahmin = ka_reg.predict(X_test)
mse_ka = MSE(y_tahmin, y_test)
rmse_ka = mse_ka**(1/2)

# Print rmse_dt
print("Karar Ağacının RMSE değeri : {:.2f}".format(rmse_ka))
```

**Parametre Ayarlama**

* Regresyon ağaçları oluşturuken aşağıdaki parametreleri ayarlayarak modelinizin performansını değiştirebilirsiniz.

* max_depth : Ağacın kaç seviye aşağıya ineceğini gösterir. 1 ila 32 arasında değişir. Yüksek değerler overfittinge yol açabilir.

* min_samples_split : Bir düğümün dallanması için en az kaç örnekleme ihtiyaç duyulduğunu gösterir.

* min_samples_leaf : min_samples_split'e benzemektedir. Oluşturulacak yeni düğümdeki minimum örneklem sayısını gösterir.

* max_features : Yeni düğümler oluşturuken göz önüne alınması gereken değişken sayısını gösterir.
